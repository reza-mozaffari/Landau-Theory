# Landau Theory

$$f = \frac{1}{2}a_0(T-T_c)m^2 + \frac{1}{4}bm^4 - hm$$

$$f = b\left[\frac{1}{2}\frac{a_0(T-T_c)}{b}m^2 + \frac{1}{4}m^4 - \frac{h}{b}m\right]$$

$$f = b\hat{f},\quad \tau = \frac{a_0(T-T_c)}{b},\quad \hat{h} = \frac{h}{b}$$

$$\hat{f} = \frac{1}{2}\tau m^2 + \frac{1}{4}m^4 - \hat{h}m$$

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
def f(x, tau, h):
    return 0.5*tau*x**2 + 0.25*x**4 - h*x
```

```python
x = np.linspace(-2.5, 2.5, 300)
plt.plot(x, f(x, 0.75, 0.0))
plt.plot(x, f(x, 0.0, 0.0))
plt.plot(x, f(x,-0.75, 0.0))
plt.xlabel("$m$", fontsize=18)
plt.ylabel("$\hat{f}$", fontsize=18)
plt.axhline(y=0, color="gray", linestyle="--")
plt.axvline(x=0, color="gray", linestyle="--")
plt.ylim(-0.25,1)
plt.xlim(-1.75, 1.75)
plt.show()
```

```python
x = np.linspace(-2.5, 2.5, 300)
plt.plot(x, f(x, 0.75, 0.0))
plt.plot(x, f(x, 0.75, 0.5))
plt.plot(x, f(x, 0.75, 1.0))
plt.ylim(-1.0,1)
plt.xlim(-2.25, 2.25)
plt.xlabel("$m$", fontsize=18)
plt.ylabel("$\hat{f}$", fontsize=18)
plt.axhline(y=0, color="gray", linestyle="--")
plt.axvline(x=0, color="gray", linestyle="--")
plt.show()
```

```python
x = np.linspace(-2.5, 2.5, 300)
plt.plot(x, f(x, 0.0, 0.0))
plt.plot(x, f(x, 0.0, 0.5))
plt.plot(x, f(x, 0.0, 1.0))
plt.ylim(-1.0,1)
plt.xlim(-2.25, 2.25)
plt.xlabel("$m$", fontsize=18)
plt.ylabel("$\hat{f}$", fontsize=18)
plt.axhline(y=0, color="gray", linestyle="--")
plt.axvline(x=0, color="gray", linestyle="--")
plt.show()
```

```python
x = np.linspace(-2.5, 2.5, 300)
plt.plot(x, f(x,-0.75, 0.0))
plt.plot(x, f(x,-0.75, 0.5))
plt.plot(x, f(x,-0.75, 1.0))
plt.ylim(-1.25,1)
plt.xlim(-2.25, 2.25)
plt.xlabel("$m$", fontsize=18)
plt.ylabel("$\hat{f}$", fontsize=18)
plt.axhline(y=0, color="gray", linestyle="--")
plt.axvline(x=0, color="gray", linestyle="--")
plt.show()
```

```python
from scipy import optimize

def df(x, tau, h):
    return tau*x + x**3 - h

def df2(x, tau, h):
    return tau + 3.0*x**2
```

```python
rootp = []
tau = []
for i in range(0,1000):
    tau.append(5.0-i*0.01)
    rootp.append(optimize.newton(df, 2.5, fprime=df2, args=(tau[i], 0.0)))

plt.plot(tau, rootp)
plt.axhline(y=0, color="gray", linestyle="--")
plt.axvline(x=0, color="gray", linestyle="--")
plt.show()
```

```python
rootp = []
rootph = []
tau = []
for i in range(0,1000):
    tau.append(5.0-i*0.01)
    rootph.append(optimize.newton(df, 2.5, fprime=df2, args=(tau[i], 0.25)))
    rootp.append(optimize.newton(df, 2.5, fprime=df2, args=(tau[i], 0.0)))

plt.plot(tau, rootp)
plt.plot(tau, rootph)
plt.axhline(y=0, color="gray", linestyle="--")
plt.axvline(x=0, color="gray", linestyle="--")
plt.show()
```
